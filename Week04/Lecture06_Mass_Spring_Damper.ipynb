### Obtener el modelo para el sistema Masa - Resorte - Amortiguador
<br>
<div align='justify'>A continuación se presenta el esquema de una masa $m$ unida a un amortiguador y a un resorte, que están conectados en paralelo. A su vez, sobre la masa se ejerce una fuerza $F$, que puede o no depender del tiempo: </div> <br>

<img src=https://raw.githubusercontent.com/juhincapiem/ModMat/refs/heads/main/Diagrams/MasaResorteAmortiguador.jpg alt="Alternative text" width="400" />


<div align='justify'> Sigamos los pasos de la siguiente metodología para plantear el modelo: </div> <br>


1. Asumir la dirección de movimiento. Para este caso, se asumen que la masa $m$ empieza a moverse hacia la derecha.
<br>
1. Plantear el Diagrama de Cuerpo Libre. Toda fuerza que vaya en la misma dirección de la dirección de movimiento que se asumió, es positiva. 
<br>
1. Aplicar la segunda Ley de Newton: $\sum F = m\cdot a$.
<br>

La ecuación que se debe obtener es:

$$
    ma = -kx - cv + F
$$

Donde $k$ es la constante de resorte y $c$, en algunos textos se pone como $\beta$, es la constante de amortiguamiento.

### Tipos de comportamiento
<br>
<div align='justify'> Sabemos que existen 3 diferentes respuestas (sobre amortiguado, críticamente amortiguado y sub-amortiguado) de un sistemas mecánica compuesto por masa, resorte y amortiguador cuando es perturbado. Hay distintas formas de anticiparnos a estas respuestas. Revisemos una de ellas: </div> 
<br>

1. Proponer un cambio de variable.
$$
    X_1 = x \, \, \text{y} \, \,  X_2 = v
$$
1. Plantear dos ecauciones de evolución.
<br>
$$
\begin{aligned}
\dot{X}_2 &=-\frac{c}{m} X_2-\frac{k}{m} X_1+F \\
\dot{X}_1 &=X_2
\end{aligned}
$$
1. Presentarlo en forma matricial.
<br>
$$
\left[\begin{array}{l}
\dot{x}_1 \\
\dot{x}_2
\end{array}\right]=\left[\begin{array}{cc}
0 & 1 \\
-\frac{k}{m} & -\frac{c}{m}
\end{array}\right]\left[\begin{array}{l}
x_1 \\
x_2
\end{array}\right]+\left[\begin{array}{l}
0 \\
F
\end{array}\right]
$$
1. Encontrar los autovalores para la matriz de coeficientes.
<br>
$$
|A-I\lambda| = 0
$$
<br>
Donde la matriz de coeficientes es:
$$
A = \left[\begin{array}{cc}
0 & 1 \\
-\frac{k}{m} & -\frac{c}{m}
\end{array}\right]
$$
1. Clasificar el comportamiento:
<br>
$$
\begin{array}{l|l}
\textbf { Tipo de respuesta } & \textbf { Autovalores } \\
\hline \text { Sobreamortiguado } & \text {  Reales y diferentes } \\
\hline \text { Críticamente amortiguado } & \text { Reales e iguales } \\
\hline \text { Subamortiguado } & \text { Complejos }
\end{array}
$$

### Solución analítica para la ecuación homogénea
<br>
<div align='justify'> Como son tres tipos de respuesta, esperamos que hayan tres tipos de soluciones analíticas. Para encontrar la respectiva solución, nos conviene darle una forma "especial" a nuestro modelos de segundo orden:</div> 
<br>

$$
    a +2 \gamma v +\omega^2 x= F
$$

Donde $2 \gamma = \frac{c}{m}$ y $\omega^2 = \frac{k}{m}$

Teniéndolo de esta forma, los autovalores se pueden calcular usando las siguientes expresiones:

$$
\lambda_1=-\gamma+\sqrt{\gamma^2-\omega^2}, \quad \lambda_2=-\gamma-\sqrt{\gamma^2-\omega^2}
$$
<br><br>
#### Sobreamortiguado

$$
x(t)=e^{-\gamma t}\left(c_1 e^{\sqrt{\gamma^2-\omega^2} t}+c_2 e^{-\sqrt{\gamma^2-\omega^2} t}\right)
$$

Los coeficientes se calculan de la siguiente manera:

$$
C_2= \frac{x\prime(0) - \lambda_1}{\lambda_2 - \lambda_1}, \quad C_1=x(0) - C_2
$$

<br><br>
#### Críticamente amortiguado

$$
x(t)=e^{-\gamma t}\left(c_1+c_2 t\right)
$$

Los coeficientes se calculan de la siguiente manera:

$$
C_1= x(0), \quad C_2=x\prime(0) + \gamma C_1
$$

<br><br>
#### Subamortiguado

$$
x(t)=e^{-\gamma t}\left(c_1 \cos \left(\sqrt{\omega^2-\gamma^2} t\right)+c_2 \operatorname{sen} \left(\sqrt{\omega^2-\gamma^2} t\right)\right)
$$

Los coeficientes se calculan de la siguiente manera:

$$
C_1= x(0), \quad C_2= \frac{x\prime(0) + \gamma C_1}{\sqrt{\omega^2 - \gamma^2}}
$$
