### Introducción
1. Recordar el modelo básico de vaciado de tanque:

$$
    \frac{dh}{dt} = \frac{-Q}{A}
$$

1. ¿Sabe de dónde sale este modelo? ¿Qué suposiciones se hicieron? Sale del Teorema de Transporte de Reynolds

1. ¿Cómo varía el cauda de salida?

$$
    Q = a\sqrt{2gh}
$$

1. ¿Cuál sería la solución analítica al modelo?

$$

$$
