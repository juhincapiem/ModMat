### Introducción

<div align='justify'>El modelo de vaciado de tanque, al igual que el modelo compartimental de tanques interconectados, proviene del Teorema de Transporte de Reynolds empleado en la conservación de la masa. </div><br>

1. ¿Qué dice el TTR para la conservación de la masa?
<br>
$$
\frac{D M_{\text {sis }}}{D t}=\frac{\partial}{\partial t} \int_{\mathrm{vc}} \rho  d \forall+\int_{\mathrm{sc}} \rho  \mathbf{V} \cdot \hat{\mathbf{n}} d A \tag{1}
$$
<br>

1. Podemos hacer tres simplificaciones: 
<br>
$$
\frac{D M_{\text {sis }}}{D t} = 0
$$ 
<br>
$$
\frac{\partial}{\partial t} \int_{\mathrm{vc}} \rho  d \forall = \frac{dm_{vc}}{dt}
$$
<br>
$$
\int_{\mathrm{sc}} \rho  \mathbf{V} \cdot \hat{\mathbf{n}} d A = \sum \dot m_{salen} - \sum \dot m_{entran}
$$
<br>
1. Actualizamos la ecuación 1:
<br>
$$
    0 = \frac{dm_{vc}}{dt} + \sum \dot m_{salen} - \sum \dot m_{entran}
$$
<br>
Y si despejamos el término $\frac{dm_{vc}}{dt}$:
<br><br>
$$
    \frac{dm_{vc}}{dt} =  \sum \dot m_{entran} - \sum \dot m_{salen} \tag{2}
$$
<br>

<div align='justify'>En pocas palabras, la ecuación 2 indica que "El cambio en el tiempo de la masa contenida en el Volumen de Control (VC) es igual a la diferencia entre el total de flujo másico que entró y el total de flujo másico que salió". </div><br>

<div class="alert alert-block alert-warning">
<b> ¿Esto cómo lo puedo usar para obtener el modelo para vaciado de tanque?</b>
</div>    
<br>

<div align='justify'> En primer lugar es necesario conocer la geometría del tanque al que le vamos a "fabricar" el modelo de vaciado, ¿Cómo es?</div><br>


<img src=https://raw.githubusercontent.com/juhincapiem/ModMat/refs/heads/main/Diagrams/drenadoTanque.png alt="Esquema de vaciado de tanque" width="300" />


1. Para este caso particular podemos expresar la ecuación 2 como:
<br>
$$
     \frac{d (\rho \forall)}{dt} = - \dot m_{sale} \tag{3}
$$
<br>
1. El volumen del tanque se puede expresar como la multiplicación entre el área transversal y la altura: $A_t \times h$, y el flujo másico como $\rho \dot \forall_{sale}$. Reemplazando en la ecuación 3:
<br><br>
$$
     \frac{d (\rho A_t h)}{dt} = - \rho \dot \forall_{sale}
$$
<br>
1. ¿La densidad, área transversal y la altura se pueden considerar constantes?
<br><br>
$$
    \frac{dh}{dt} = \frac{-\dot \forall_{sale}}{A_t}
$$
<br>
1. ¿Qué nos falta para poder resolver este modelo? ¡Conocer el caudal de salida! 
<br><br>
$$
    \dot \forall_{sale} = a\sqrt{2gh}
$$
<br>
Donde $a$ es el área del agujero, $g$ es la gravedad y $h$ es la altura.
<div class="alert alert-block alert-info">
<b> ¿Recuerdan cómo se obtiene la expresión $\sqrt{2gh}$? ¿Y qué simplificaciones se asumieron para obtenerla?</b>
</div>    
1. Al juntarlo obtenemos:
<br><br>
$$
    \frac{dh}{dt} = \frac{-a\sqrt{2gh}}{A_t}
$$
<br>
1. ¿Cuál sería la solución analítica al modelo?
<br><br>
$$
\mathbf{h(t)}=\left(\sqrt{\mathbf{h}_{0}}\ -{\frac{\mathbf{a}\cdot{\sqrt{2\ {\mathrm{g}}}}}{2\cdot{\mathbf{A}}}}\ \mathbf{t}\right)^{2} \tag{4}
$$
<br>
1. Si quisieramos ser un poco más precisos, podríamos añadir un coeficiente $C_c$ para corregir el área de descarga y un coeficiente $C_v$ que corrige la velocidad de salida:
<br><br>
$$
\mathbf{h(t)}=\left(\sqrt{\mathbf{h}_{0}}\ -{\frac{\mathbf{C}_{\mathrm{c}}\cdot C_{\mathrm{v}}\cdot\mathbf{a}\cdot{\sqrt{2\ {\mathrm{g}}}}}{2\cdot{\mathbf{A}}}}\ \mathbf{t}\right)^{2} \tag{5}
$$
<br>
El coeficiente de descarga $C_D$ o simplemente $C$ (para no confundirlo con el coeficiente de arrastre) es el producto de la multiplicación entre $C_c$ y $C_v$. Algunos valores se presentan a continuación:

$$
\mathrm{RMSD}=\sqrt{\frac{\sum_{i=1}^N\left(x_i-\hat{x}_i\right)^2}{N}} \\
$$

$\text{RMSD significa raiz del error cuadrático medio}$.

$i=\text { variable i }$

$N =\text { número total de datos }$

$x_i=\text {valores reales }$

$\hat{x}_i \quad=\text { valores aproximados}$

<div class="alert alert-block alert-success">
<b> Paso 03: Crear una función en donde se resuelva la ecuación 4 de vaciado de tanque. La función debe aceptar un vector de tiempo, el paso temporal y los parámetros del modelo. </b>
</div><br>  

```python
# Problema 0: definir función para evaluar la pendiente
pendienteH = lambda g,h,A0,At,Cd: -Cd*A0*np.sqrt(2*g*h)/At

# Problema 1: definir función para usar Euler
def eulerForward(tiempo, dt, info):
    g = info[0]
    h0 = info[1]
    A0 = info[2]
    At = info[3]
    Cd = info[4]

    hNum = np.zeros(len(tiempo))
    hNum[0] = h0
    for i in range(0, len(tiempo)-1):
        hNum[i+1] = hNum[i] + pendienteH(g,hNum[i],A0, At, Cd)*dt
    return hNum
```

```python
# Problema 2: parámetros de control temporal y vector del tiempo
t0 = 0.0; tf = 60.0; dt = 0.001
tiempo = np.arange(t0, tf +dt, dt)

# Problema 3: demás parámetros del problema
g = 9.81*(1000/1); h0 = 350; A0 = np.pi*6**2/4; At = np.pi*90**2/4; Cd = 1.0
info = np.array([g, h0, A0, At, Cd])

# Problema 4: llamar la función Euler
hNum = eulerForward(tiempo, dt, info)
```

<div class="alert alert-block alert-success">
<b> Modelo Legendario </b>
</div><br> 


$$
    h \left(g+ \frac{d^2h}{dt^2} \right) = \frac{1}{2}\left(\frac{dh}{dt} \right)^2 \left[ \left(\frac{A}{a}\right)^2-1 \right]
$$

<div class="alert alert-block alert-success">
<b> Modelo semi-legendario </b>
</div><br> 

¿Cómo se puede resovler este modelo? Una opción es suponer que $g >> \frac{d^2h}{dt^2}$, y por tanto, se desprecia la derivada de segundo orden:

$$
    h g = \frac{1}{2}\left(\frac{dh}{dt} \right)^2 \left[ \left(\frac{A}{a}\right)^2-1 \right]
$$

Si se despeja $\frac{dh}{dt}$ se obtiene la siguiente ecuación:

$$
  \frac{dh}{dt} = -\left\{ \frac{2hg}{\left[ \left(\frac{A}{a}\right)^2-1 \right]}\right\}^{1/2}
$$




```python
t0 = 0.0; tf = 60.0; dt = 0.001
tiempoSL = np.arange(t0,tf + dt,dt)
hNumSL = np.zeros(len(tiempoSL))

# Condiciones iniciales 
hNumSL[0] = 350

for i in range(0,len(tiempoSL)-1):
    hNumSL[i+1] = hNumSL[i] + -np.sqrt(2*hNumSL[i]*g)/np.sqrt((At/A0)**2-1)*dt
```

Si aún así no está convencido con los anteriores modelos, se puede resolver el modelo completo:

$$
    h \left(g+ \frac{d^2h}{dt^2} \right) = \frac{1}{2}\left(\frac{dh}{dt} \right)^2 \left[ \left(\frac{A}{a}\right)^2-1 \right]
$$

```python
t0 = 0.0; tf = 59.0; dt = 0.00001
tiempoL = np.arange(t0,tf + dt,dt)
VNum = np.zeros(len(tiempoL))
hNumL = np.zeros(len(tiempoL))

# Condiciones iniciales 
hNumL[0] = 350
VNum[0] = 0

for i in range(0,len(tiempoL)-1):
    VNum[i+1] = ((0.5*(VNum[i])**2*((At/A0)**2 - 1) )/hNumL[i] - g)*dt + VNum[i]
    hNumL[i+1] = VNum[i]*dt + hNumL[i]
    
    
```

<div class="alert alert-block alert-danger">
<b> Bonus. Existe una manera más exacta de hallar el coeficiente de descarga: haciendo una regresión lineal de un modelo potencial</b>
</div><br>  

Pero... ¿Cómo se hace una regresión lineal de un modelo potencial? Aquí va:

Podemos plantear una relación del tipo pontencial entre la altura $h$ y el caudal $\dot \forall$

$$
\dot \forall(h) = \delta\,h^{n}
$$

donde $\delta$ y $n$ son el coeficiente de proporcionalidad y la potencia del modelo, respectivamente.

Para poder determinar los valores de $\delta$ y $n$ dado un conjunto de valores experimentales para $h$, podemos hacer regresión lineal haciendo un pequeño ajuste del modelo.  Para tal fin se toman los logaritmos naturales (o cualquier otro logaritmo) de cada lado del modelo de potencia, con lo que se obtiene:

\begin{equation}
\begin{aligned}
\dot \forall &= \delta\,h^{n} \\
\log\left(\dot \forall\right) &= \log\left(\delta\,h^{n}\right)\\
Y &= \log{\delta} + \log{h^{n}}\\
Y &= \log{\delta} + n\,\log{h}\\
\end{aligned}
\end{equation}

Si la ponemos de la siguiente forma:
$$
Y = n\,\log{h} + \log{\delta}  \\
$$


¿No se les parece a algo muy conocido? ¡Tiene la misma forma que una función lineal!

$$
Y = M X + B
$$


Así pues, encontrar el coeficiente $\delta$ y el exponente $n$, es simplemente un proceso de determinar el intersecto con el eje $Y$ y la pendiente de la función.

Supongamos que tenemos datos experimentales de caudal:
