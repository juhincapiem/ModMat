También podemos crear matrices usando la misma función **np.array**. Miremos cómo se hace:

```python
M = np.array([[3,4,-1],[2,0,1],[1,3,-2]])
print(M)
print(M[-1,-1])
```

¿Qué operaciones nos permite hacer Python?

```python
M_inv = np.linalg.inv(M)
print(M_inv)

#¿Cómo verificamos que es la inversa?
print(np.dot(M, M_inv))
```

```python
M_tr = np.transpose(M)
print(M_tr)
#¿Cómo verificamos que es la transpuesta?
```

\begin{align*}
    & \frac{384}{17}x + \frac{240}{17}y = 8 \times 12 \\

### Gráficas en Python: matplotlib

<div align='justify'> Matplotlib es una súper biblioteca para crear visualizaciones estáticas, animadas e interactivas en Python. Es bastante poderoso; muchos investigadores las usan para hacer </div>

https://matplotlib.org/

```python
import matplotlib.pyplot as plt
```

```python
# make a stream function:
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
# make U and V out of the streamfunction:
V = np.diff(Z[1:, :], axis=1)
U = -np.diff(Z[:, 1:], axis=0)

# plot:
fig, ax = plt.subplots()

ax.streamplot(X[1:, 1:], Y[1:, 1:], U, V)

plt.show()
```

### Errores 
<div align='justify'> Conocer el error de nuestras estimaciones responde a la pregunta: ¿Qué tan bueno es nuestro modelo? ¿O qué tan malo? Y ¿Para qué queremos saber eso? ¿De qué nos sirve?</div>

**Error absoluto total:** valor aproximado - valor verdadero

**Error relativo:** Error absoluto total/valor verdadero

**Por ejemplo:** Nos piden montar un sensor de temperatura (amarillo) en un cuarto, y para esto nos dan las coordenadas, como se pueden ver en la imagen de abajo. Sin emabrgo, por cosas de la vida, no se pudo poner en el lugar exacto. Además, el sensor llegó un poco "maluco", así que no arroja muy buenas mediciones.

 En este montaje vemos muchas cosas que están mal ¿Cómo cuantificarlas? Usando el concepto de error. Vamos a fraccionarlo:
 
 1. $x$ es el valor verdadero de entrada. ¿En este ejemplo cuál sería?
 1. $f(x)$ es el valor verdadero de salida, el que buscamos, porque es la función exacta. ¿Cuál es?
 1. $x^*$ es el valor de entrada aproximado. ¿Cuál es?
 1. $f^*(x^*)$ es la función aproximada evaluada con el valor aproximado. ¿Cuál es?
 
 $$
 \textbf{Error total} = f^*(x^*) - f(x)
 $$
